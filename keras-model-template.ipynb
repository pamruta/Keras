```python
import keras
from keras import layers
```

```python
from keras.datasets import fashion_mnist
(xtrain, ytrain), (xtest, ytest) =  fashion_mnist.load_data()
```

```python
def create_seq_model():
  model = keras.Sequential()
  model.add(layers.Input(shape=(28, 28)))
  model.add(layers.Flatten())
  model.add(layers.Dense(64, activation='relu'))
  model.add(layers.Dense(64, activation='relu'))
  model.add(layers.Dense(10, activation='softmax'))
  return model
```

```python
seq_model = create_seq_model()
seq_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
seq_model.fit(xtrain, ytrain, epochs=10)
```

```python
def create_fun_model():
  inputs = keras.Input(shape=(28, 28))
  features = layers.Flatten()(inputs)
  features = layers.Dense(64, activation='relu')(features)
  features = layers.Dense(64, activation='relu')(features)
  outputs = layers.Dense(10, activation='softmax')(features)

  return keras.Model(inputs=inputs, outputs=outputs)
```

```python
model = create_fun_model()
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(xtrain, ytrain, epochs=10)
```

```python
def create_conv2d_model():

  inputs = keras.Input(shape=(28, 28, 1))
  features = layers.Conv2D(16, 2, activation='relu')(inputs)
  features = layers.MaxPooling2D(2)(features)
  features = layers.Conv2D(32, 2, activation='relu')(features)
  features = layers.MaxPooling2D(2)(features)

  features = layers.Flatten()(features)
  features = layers.Dense(32, activation='relu')(features)
  outputs = layers.Dense(10, activation='softmax')(features)

  return keras.Model(inputs=inputs, outputs=outputs)
```

```python
conv2d_model = create_conv2d_model()
conv2d_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
conv2d_model.fit(xtrain, ytrain, epochs=5)
```
