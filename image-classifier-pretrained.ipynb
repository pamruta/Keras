```python
#import required libraries

import keras
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import image
```

```python
# test image
img_file = "burger.jpg"
img = image.imread(img_file)
plt.imshow(img)
plt.show()
```

```python
# test image size / resolution
img.shape
```

```python
# pre-trained models available in keras hub
import keras_hub
```

```python
# load model
# list of available models can be found at
# https://keras.io/keras_hub/presets/

# popular options for image classification include:
# resnet_50_imagenet, resnet_v2_50_imagenet, resnet_101_imagenet
# vgg_16_imagenet

classifier = keras_hub.models.ImageClassifier.from_preset("resnet_v2_101_imagenet")
```

```python
# test image must be of same size as training set on which model is pre-trained
classifier.preprocessor.image_size=(224,224)
```

```python
# predict model output on test image

predictions = classifier.predict(np.array([img]))
predictions.shape
```

```python
# print predicted classes
preds = keras_hub.utils.decode_imagenet_predictions(predictions, top=5)
preds
```
